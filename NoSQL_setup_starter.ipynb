## Part 2: Update the Database

1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
new_restaurant = {
    "name": "Penang Flavours",
    "cuisine": "Halal",
    "location": "Greenwich",
    "rating": None, # because it hasn't been rated yet
    "type": "Restaurant",
    "opening_hours": {
        "monday": "10:00-22:00",
        "tuesday": "10:00-22:00",
        "wednesday": "10:00-22:00",
        "thursday": "10:00-22:00",
        "friday": "10:00-23:00",
        "saturday": "10:00-23:00",
        "sunday": "12:00-21:00"
    },
}
```

```python
# Insert the new restaurant into the 'establishments' collection
result = db.establishments.insert_one(new_restaurant)
print("New restaurant added with _id:", result.inserted_id)
```

```python
# Check that the new restaurant was inserted

new_restaurant_check = db.establishments.find_one({"name": "Penang Flavours"})

# Check if the query returned a document
if new_restaurant_check:
    print("The new restaurant 'Penang Flavours' has been successfully inserted.")
    pprint(new_restaurant_check)  # print the document details
else:
    print("The new restaurant 'Penang Flavours' was not found in the database.")
```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Define the query to search for the BusinessType
query = {"BusinessType": "Restaurant/Cafe/Canteen"}

# Define the projection to specify only the BusinessTypeID and BusinessType fields to return
projection = {"BusinessTypeID": 1, "BusinessType": 1, "_id": 0}

# Find all documents matching the query
business_type_documents = db.establishments.find(query, projection)

# Iterate through the cursor and print each document
for document in business_type_documents:
    print(document)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python

correct_business_type_id = 1

restaurant_filter = {"name": "Penang Flavours"}

update_data = {"$set": {"BusinessTypeID": correct_business_type_id}}

result = db.establishments.update_one(restaurant_filter, update_data)

if result.matched_count > 0:
    print(f"Restaurant 'Penang Flavours' has been updated with BusinessTypeID: {correct_business_type_id}")
else:
    print("Restaurant 'Penang Flavours' was not found in the database.")
```

```python
# Confirm that the new restaurant was updated
# Query to find the "Penang Flavours" restaurant
restaurant_query = {"name": "Penang Flavours"}

# Execute the query
updated_restaurant = db.establishments.find_one(restaurant_query)

# Check if the restaurant was found and print the result
if updated_restaurant:
    print("Updated restaurant details:")
    print(updated_restaurant)
else:
    print("Restaurant 'Penang Flavours' was not found in the database.")
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"
# Define the query to count documents with LocalAuthorityName as "Dover"
query = {"LocalAuthorityName": "Dover"}

# Count the documents
dover_count = db.establishments.count_documents(query)

# Print the counting resault 
print("Number of documents with LocalAuthorityName as 'Dover':", dover_count)
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
# Define the query to match documents where LocalAuthorityName is "Dover"
query = {"LocalAuthorityName": "Dover"}

# Delete all matching documents
delete_result = db.establishments.delete_many(query)

# Print the count of deleted documents
print("Number of documents deleted:", delete_result.deleted_count)
```

```python
# Check if any remaining documents include Dover
# Define the query to find documents with LocalAuthorityName as "Dover"
query = {"LocalAuthorityName": "Dover"}

# Count the documents
remaining_dover_count = db.establishments.count_documents(query)

# Check and print whether there are any remaining documents
if remaining_dover_count == 0:
    print("There are no remaining documents with LocalAuthorityName as 'Dover'.")
else:
    print(f"There are still {remaining_dover_count} documents with LocalAuthorityName as 'Dover'.")
```

```python
# Check that other documents remain with 'find_one'
# Execute the find_one method to get a single document from the collection
sample_document = db.establishments.find_one()

# Check if a document was found and print the result
if sample_document:
    print("A document from the 'establishments' collection:")
    print(sample_document)
else:
    pprint("No documents found in the 'establishments' collection.")
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
establishments.update_many({}, [{'$set':{"geocode.latitude":{"$toDecimal": "$geocode.latitude"}, 
                                          "geocode.longitude":{"$toDecimal": "$geocode.longitude"}}}])
```

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [{'$set':{ "RatingValue" : None}}])
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many({}, [{'$set':{"RatingValue":{"$toInt": "$RatingValue"}}}])
```

```python
# Check that the coordinates and rating value are now numbers

pprint(establishments.find_one())
```
